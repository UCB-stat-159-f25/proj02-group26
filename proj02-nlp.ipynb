````python
import spacy
import pandas as pd

try:
    nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
except OSError:
    print("ERROR: spaCy model 'en_core_web_sm' not found. Please run: python -m spacy download en_core_web_sm")
    pass

#clean text
def clean_text_spacy(text):
    """
    Tokenizes, lemmatizes, removes stop words and punctuation, and returns 
    a single string of cleaned lemmas for a given text.
    """
    if pd.isna(text) or text is None:
        return ""

    # process text with the spaCy
    doc = nlp(str(text))

    # Filter for tokens and extract the lemma
    clean_lemmas = [
        token.lemma_.lower()  # Use the base form (lemma) and lowercase it
        for token in doc
        if token.is_alpha          # Keep only words (no numbers or symbols)
        and not token.is_punct      # removes punctuation
        and not token.is_stop       # Remove common English stop words
    ]

    return " ".join(clean_lemmas)

print("Starting NLP processing...")
df['cleaned_text'] = df['text'].apply(clean_text_spacy)
print("Processing complete!")

#Verification
print("\n--- Verification Sample (First 5 speeches) ---")
print(df[['year', 'text', 'cleaned_text']].head())
````

### Analyze Tokens vs Lemmas

#### Token List

Create a list of tokens across all speeches that are not spaces, stopwords, or punctuation. Make each token lowercase as well. *Hint: each element of the list we just created are themselves lists of tokens. Token objects have attributes `is_stop`, `is_punct`, and `is_space`.*

````python
# print top 20 tokens
# Hint - use Counter, and one of the Counter object's methods to display the top 20d
````

````python
import itertools

def get_filtered_tokens(text, nlp_model):
    """
    Processes text to return a list of tokens, excluding stop words, 
    punctuation, and spaces, and converts to lowercase.
    """
    if pd.isna(text) or text is None:
        return []
        
    # process using the spaCy model
    doc = nlp_model(str(text))
    
    # filter tokens and return lowercase text of the token
    filtered_tokens = [
        token.text.lower()
        for token in doc
        if not token.is_space
        and not token.is_punct
        and not token.is_stop
    ]
    return filtered_tokens

print("Processing speeches to extract filtered tokens...")

list_of_token_lists = df['text'].apply(lambda x: get_filtered_tokens(x, nlp)).tolist()
all_clean_tokens = list(itertools.chain.from_iterable(list_of_token_lists))


# fet frequencies
token_counts = Counter(all_clean_tokens)

# top 20 most common tokens
print("Processing complete!")
print("\n--- Top 20 Most Frequent Tokens ---")
print(token_counts.most_common(20))
````

#### Lemma List

Do the same as above, but for lemmas. *Hint: recall lemmas are components of words. Each token should have an attribute to extract the lemma.*

````python
import pandas as pd
from collections import Counter
import itertools

def get_filtered_lemmas(text, nlp_model):
    """
    Processes text to return a list of lemmas, excluding stop words, 
    punctuation, and spaces, and converts to lowercase.
    """
    if pd.isna(text) or text is None:
        return []
        
    doc = nlp_model(str(text))
    
    # Filter tokens
    filtered_lemmas = [
        token.lemma_.lower() # <-- Key difference: using .lemma_
        for token in doc
        if not token.is_space
        and not token.is_punct
        and not token.is_stop
    ]
    return filtered_lemmas
    
print("Processing speeches to extract filtered lemmas...")

list_of_lemma_lists = df['text'].apply(lambda x: get_filtered_lemmas(x, nlp)).tolist()

all_clean_lemmas = list(itertools.chain.from_iterable(list_of_lemma_lists))
lemma_counts = Counter(all_clean_lemmas)

# print the top 20 most common lemmas
print("Processing complete!")
print("\n--- Top 20 Most Frequent Lemmas (Base Forms) ---")
print(lemma_counts.most_common(20))
````

#### Token versus Lemma Comparison

What do you notice about the top tokens versus the top lemmas? 
Consider two tokens - "year" and "years" - how do their counts compare to the lemma "year"?
What about the lemma "child"?

The overall difference between the top tokens and lemmas is that the Lemma List provides a more concise and accurate measure of the speeches, while the Token List scatters the counts across different forms of the word.

Tokens (raw words) have their counts divided by different versions of the word, such as "job" and "jobs." This separation causes the frequency of any single form of that word to accumulate less total count.

Lemmas represent all the variant forms of the word, and are mapped into a single base word (the lemma). This combines the counts, leading to higher totals and better indications of the more important concepts in the corpus.

The lemma 'child' has a frequency count of 332. Since the lemma 'child' is the base form, the count represents the times the concept of 'child' appears in the text, which can include both the singular token 'child' and the frequent plural token 'children.' The high count represents the significance of the mentioned topics related to children, family, and education in modern State of the Union addresses.

#### Common Words per Year Function

Fill in the below function to obtain the n-most common words in speeches for a given year.

inputs: 
- df raw unprocessed sou dataframe
- year
- n
outputs: 
- top n words for that years

steps:
- subset the dataframe for the year of interest - note the years might not be in int type
- process the subsetted dataframe with spacy
- get the lemmas across all those speeches
- count the top n lemmas

````python
# test it on 2024
````

````python
def get_common_words_per_year(df: pd.DataFrame, year, n: int) -> list:
    numeric_years = pd.to_numeric(df['year'], errors='coerce')
    subset_df = df[numeric_years == int(year)].copy()

    if subset_df.empty:
        print(f"No speeches found for the year: {year}. Please verify the year is present in the DataFrame.")
        return []
    
    if 'text' not in subset_df.columns or subset_df['text'].isnull().all():
        print("Error: 'text' column missing or empty in the subset.")
        return []

    def process_and_get_lemmas(text):
        if pd.isna(text):
            return []
        
        doc = nlp(str(text))
    
        filtered_lemmas = [
            token.lemma_.lower()
            for token in doc
            if not token.is_space
            and not token.is_punct
            and not token.is_stop
        ]
        return filtered_lemmas

    subset_df['lemmas'] = subset_df['text'].apply(process_and_get_lemmas)
    all_lemmas = list(itertools.chain.from_iterable(subset_df['lemmas']))
    
    if not all_lemmas:
        return []

    # top n lemmas
    lemma_counts = Counter(all_lemmas)
    
    return lemma_counts.most_common(n)
````

````python
# Assuming the function get_common_words_per_year(df, year, n) is defined
# and df (DataFrame) and nlp (spaCy model) are available.

# Call 1: Analyze the most recent year (2024)
year_2024 = 2024
n_words = 10
top_2024_lemmas = get_common_words_per_year(df, year_2024, n_words)

print(f"--- Top {n_words} Lemmas in {year_2024} ---")
print(top_2024_lemmas)


# Call 2: Analyze a past year, e.g., 2005 (George W. Bush)
year_2005 = 2005
top_2005_lemmas = get_common_words_per_year(df, year_2005, n_words)

print(f"\n--- Top {n_words} Lemmas in {year_2005} ---")
print(top_2005_lemmas)
````
